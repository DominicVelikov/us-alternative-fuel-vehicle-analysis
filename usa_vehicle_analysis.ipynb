```python
import pandas as pd

df = pd.read_csv('Vehicle Data.csv')
```

```python
df.head(10)
```

```python
df = df.drop(index=8)

df.head(10)
```

```python
df.info() 
```

```python
# set object rows to int by removing commas and changing data type

cols_to_convert = df.columns.drop("State")

df[cols_to_convert] = (
    df[cols_to_convert]
    .replace(",", "", regex=True)
    .apply(pd.to_numeric, errors="coerce")
    .astype("Int64")
)

df.info()
```

```python
df.describe(include='all') 
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# set all columns to snake case

df.columns = df.columns.str.lower() # sets all columns to lowercase
df.columns = df.columns.str.replace(' ','_') # replaces all spaces with underscores


df.columns = df.columns.str.replace(r"\s*\(.*", "", regex=True) # removes everything after parentheses in column names
df.columns = df.columns.str.replace(r"_$", "", regex=True) # removes underscores *that are at the end of some columns* using $ 


df.head()
```

```python
df = df.drop(columns=["unknown_fuel"])
```

```python
df["methanol"].sum() 
```

```python
df = df.drop(columns=["methanol"])

df.head()
```

## Cleaning Complete

### Now, I'll add in some columns that will aid in analysis down the line.

```python
# adding a "total column"

fuel_cols = df.columns.drop("state")

df["total"] = df[fuel_cols].sum(axis=1)

df.head()
```

```python
# adds alternative and traditional fuel columns - useful for traditional vs alternative fuel comp

alt_fuel_cols = [
    "electric",
    "plug-in_hybrid_electric",
    "hybrid_electric",
    "biodiesel",
    "ethanol/flex",
    "compressed_natural_gas",
    "propane",
    "hydrogen"
]

df["alternative_fuel_total"] = df[alt_fuel_cols].sum(axis=1)



df["traditional_fuel_total"] = df[["gasoline", "diesel"]].sum(axis=1)


electrified_cols = [
    "electric",
    "plug-in_hybrid_electric",
    "hybrid_electric"
]

df["electrified_total"] = df[electrified_cols].sum(axis=1)


df.head()
```

```python
# adds some % columns for an unbiased comparison (some states are larger than others. It's best to compare %'s rather than totals)

# alternative %
df["alternative_fuel_percent"] = (
    df["alternative_fuel_total"] / df["total"] * 100
)

# traditional %
df["traditional_fuel_percent"] = (
    df["traditional_fuel_total"] / df["total"] * 100
)

# electric %
df["electric_percent"] = df["electric"] / df["total"] * 100

# plug-in hybrid electric %
df["plug-in_hybrid_electric_percent"] = df["plug-in_hybrid_electric"] / df["total"] * 100

# hybrid electric %
df["hybrid_electric_percent"] = df["hybrid_electric"] / df["total"] * 100


# electrified %
df["electrified_percent"] = (
    df["electrified_total"] / df["total"] * 100
)

# gasoline %
df["gasoline_percent"] = df["gasoline"] / df["total"] * 100

# Diesel %
df["diesel_percent"] = df["diesel"] / df["total"] * 100


df.head()
```

```python
# dominant alternative fuel type

df["dominant_alternative_fuel_type"] = df[alt_fuel_cols].idxmax(axis=1)


df.head()
```

```python
# final fheck

df.info()
```
